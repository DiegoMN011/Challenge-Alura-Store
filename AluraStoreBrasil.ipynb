### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
loja.head()
```

#1. Análise do faturamento



```python

faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print("Faturamento Loja 1:", faturamento_loja1)
print("Faturamento Loja 2:", faturamento_loja2)
print("Faturamento Loja 3:", faturamento_loja3)
print("Faturamento Loja 4:", faturamento_loja4)
```

# 2. Vendas por Categoria

```python
print("Categorias Loja 1")
print(loja['Categoria do Produto'].value_counts())

print("\nCategorias Loja 2")
print(loja2['Categoria do Produto'].value_counts())

print("\nCategorias Loja 3")
print(loja3['Categoria do Produto'].value_counts())

print("\nCategorias Loja 4")
print(loja4['Categoria do Produto'].value_counts())
```

# 3. Média de Avaliação das Lojas


```python
print("Média Loja 1:", loja['Avaliação da compra'].mean())
print("Média Loja 2:", loja2['Avaliação da compra'].mean())
print("Média Loja 3:", loja3['Avaliação da compra'].mean())
print("Média Loja 4:", loja4['Avaliação da compra'].mean())
```

# 4. Produtos Mais e Menos Vendidos

```python
print("Loja 1")
print("Mais vendido:", loja['Produto'].value_counts().idxmax())
print("Menos vendido:", loja['Produto'].value_counts().idxmin())

print("\nLoja 2")
print("Mais vendido:", loja2['Produto'].value_counts().idxmax())
print("Menos vendido:", loja2['Produto'].value_counts().idxmin())

print("\nLoja 3")
print("Mais vendido:", loja3['Produto'].value_counts().idxmax())
print("Menos vendido:", loja3['Produto'].value_counts().idxmin())

print("\nLoja 4")
print("Mais vendido:", loja4['Produto'].value_counts().idxmax())
print("Menos vendido:", loja4['Produto'].value_counts().idxmin())
```

# 5. Frete Médio por Loja

```python
print("Frete médio Loja 1:", loja['Frete'].mean())
print("Frete médio Loja 2:", loja2['Frete'].mean())
print("Frete médio Loja 3:", loja3['Frete'].mean())
print("Frete médio Loja 4:", loja4['Frete'].mean())
```

Gráfico 1 - Faturamento


```python
import matplotlib.pyplot as plt

lojas = ['Loja 1','Loja 2','Loja 3','Loja 4']
faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.bar(lojas, faturamento)
plt.title("Faturamento por Loja")
plt.show()
```

Gráfico 2 - Avaliações

```python
avaliacoes = [
    loja['Avaliação da compra'].mean(),
    loja2['Avaliação da compra'].mean(),
    loja3['Avaliação da compra'].mean(),
    loja4['Avaliação da compra'].mean()
]

plt.bar(lojas, avaliacoes)
plt.title("Avaliação média das lojas")
plt.show()
```

Gráfico 3 - Categorias

```python
categorias = loja['Categoria do Produto'].value_counts()

plt.pie(categorias.head(5), labels=categorias.head(5).index, autopct='%1.1f%%')
plt.title("Categorias mais vendidas")
plt.show()
```
